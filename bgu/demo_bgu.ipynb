```python
import bgu
import matplotlib.pyplot as plt
```

```python
def display(highr_input, highr_output, lowr_output):
    plt.figure(figsize=(15, 10))
    plt.subplot(131)
    plt.title('High Resolution Input')
    plt.imshow(highr_input)

    plt.subplot(132)
    plt.title('Low Resolution Model Output')
    plt.imshow(lowr_output)
    
    plt.subplot(133)
    plt.title('Upsampled Output Image')
    plt.imshow(highr_output)
    
    plt.show()
```

```python
input_img_large = "Images/INPUT_IMGS/khjhvdjskd1_2_dim.png"
model_out_small = "Images/MODEL_OUTS/khjhvdjskd1_2_dim.png"

highr_input, highr_output, lowr_output = bgu.upsampler(input_img_large, model_out_small)
display(highr_input, highr_output, lowr_output)
```
