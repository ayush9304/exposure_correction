```python
import os
import cv2
import matplotlib.pyplot as plt

from bgu import bgu
from utils import enhance
```

```python
def display(highr_input, highr_output, lowr_output):
    plt.figure(figsize=(15, 10))
    plt.subplot(131)
    plt.title('High Resolution Input')
    plt.imshow(highr_input)

    plt.subplot(132)
    plt.title('Low Resolution Model Output')
    plt.imshow(lowr_output)
    
    plt.subplot(133)
    plt.title('Upsampled Output Image')
    plt.imshow(highr_output)
    
    plt.show()
```

```python
input_dir = "data/Input_Images/"
model_output_dir = "data/Model_Outputs/"
upsampled_output_dir = "data/Upsampled_Outputs/"
```

```python
enhance(input_dir, model_output_dir)
```

```python
test_img = "khjhvdjskd1_2_dim.png"

input_img_large = os.path.join(input_dir, test_img)
model_out_small = os.path.join(model_output_dir, test_img)
upsampled_out_large = os.path.join(upsampled_output_dir, test_img)

highr_input, highr_output, lowr_output = bgu.upsampler(input_img_large, model_out_small)

cv2.imwrite(upsampled_out_large, cv2.cvtColor(highr_output, cv2.COLOR_RGB2BGR))
print("Upsampled image saved to {}".format(upsampled_out_large))

display(highr_input, highr_output, lowr_output)
```
